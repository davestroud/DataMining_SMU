```python
df_test = pd.read_csv('DMProj1/test.csv', low_memory=False)
df_test.head()
```

```python
df_resources = pd.read_csv('DMProj1/resources.csv')
df_resources.head()
```

```python
df_sample_submission = pd.read_csv('DMProj1/sample_submission.csv')
df_sample_submission.head()
```

```python
#https://www.kaggle.com/codename007/a-very-extensive-end-to-end-project-donorschoose
## Merging with train and test data ##
train_resource = pd.merge(df_train, df_resources, on="id", how='left')
test_resource = pd.merge(df_test, df_resources, on="id", how='left')
```

```python
print("Size of training data : ",df_train.shape)
print("Size of test data : ",df_test.shape)
print("Size of resource data : ",df_resources.shape)
print("Size of train_resource data : ",train_resource.shape)
print("Size of test_resource data : ",test_resource.shape)
```
